```python
import pandas as pd 
import numpy as np
import chardet 
```

```python
unames = ["user_id" , "gender" , "age", "occupation" , "zip"]
users = pd.read_table("D:\\downloads\\movies lens\\data\\users.dat", sep="::",
header=None, names=unames, engine="python")

rnames = ["user_id","movie_id","rating","timestamp"]
ratings = pd.read_table("D:\\downloads\\movies lens\\data\\ratings.dat" , sep="::" , header= None , names=rnames ,engine="python" )

with open("D:\\downloads\\movies lens\\data\\movies.dat","rb") as f :
    result = chardet.detect(f.read())
mnames = ["movie_id", "title", "genres"]
Movies = pd.read_table("D:\\downloads\\movies lens\\data\\movies.dat", encoding=result["encoding"],sep="::",names=mnames,engine="python")
```

```python
data = pd.merge(pd.merge(users,ratings),Movies)
data
```

```python
data.iloc[0]
```

```python
mean_ratings = data.pivot_table("rating" , index="title"  , columns="gender",aggfunc="mean")
mean_ratings
```

```python
ratings_by_title = data.groupby("title").size()
active_title = ratings_by_title.index[ratings_by_title>=250]
active_title
```

```python
mean_ratings = mean_ratings.loc[active_title]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values("F", ascending=False)
top_female_ratings
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
mean_ratings
```

```python
sorting_by_diff = mean_ratings.sort_values("diff")
sorting_by_diff.loc[::-1]
```

```python
Movies.genres.head()
Movies['genres'] = Movies['genres'].str.split("|")
```

```python
movies_exploded = Movies.explode("genres")
movies_exploded
```

```python
rating_std_by_title = data.groupby("title")["rating"].std()
rating_std_by_title.sort_values(ascending=False).dropna()
rating_std_by_title.head()
```

```python
rating_with_genres = pd.merge(pd.merge(movies_exploded,ratings),users)
rating_with_genres
```

```python
genre_rating = rating_with_genres.groupby(['genres','age'])['rating'].mean().unstack('age')
genre_rating
```
